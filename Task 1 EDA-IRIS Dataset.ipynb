```python
#Scatter plot for Sepal Length and Sepal Width 
```

```python
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Scatterplot for Petal Length and Petal Width 
```

```python
sns.scatterplot(data=iris_df, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
#Feature Distribution of Iris Species
```

```python

features = ["SepalWidthCm", "SepalLengthCm", "PetalLengthCm", "PetalWidthCm"]

boxprops = dict(facecolor="lightblue", color="blue")
medianprops = dict(color="red")

iris_df.boxplot(
    column=features,
    by="Species",
    layout=(2, 2),
    figsize=(10, 8),
    patch_artist=True,
    boxprops=boxprops,
    medianprops=medianprops
)

plt.suptitle("")
plt.show()
```

```python
#Boxplot Analysis of Iris Features by Species
```

```python
features = ["SepalWidthCm", "SepalLengthCm", "PetalLengthCm", "PetalWidthCm"]
colors = ["skyblue", "lightgreen", "salmon", "plum"]

plt.figure(figsize=(10, 8))

for i, col in enumerate(features, 1):
    plt.subplot(2, 2, i)
    plt.hist(iris_df[col], bins=10, color=colors[i-1], edgecolor="black")
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
Conclusion & Key Insights

The visual analysis of the Iris dataset using boxplots and histograms reveals clear patterns in the distribution of sepal and petal measurements across species.

Petal features (Petal Length & Petal Width) show strong separation between species, making them highly effective for classification.

Sepal features (Sepal Length & Sepal Width) exhibit more overlap, indicating they are less reliable for distinguishing species on their own.

The histograms indicate that most features are approximately normally distributed, though some skewness is present in certain variables.

Boxplots highlight the presence of variation and potential outliers, particularly in sepal measurements.

Among the species, one group (typically Setosa) tends to have noticeably smaller petal dimensions, making it easily distinguishable from the others.
```
